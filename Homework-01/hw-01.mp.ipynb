```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import metpy.calc as mpcalc
from metpy.units import units
```

```python
# Read in radiosonde data
def read_radiosonde(fname):
    names = ['Time','P','T','U','Wsp','Wdir','Altitude','Geo_Pot']
    return pd.read_csv(fname,skiprows=[0],names=names,delimiter='\t')
```

```python
def check_calc(df):
    p = df['P'].values * units.hPa
    T = df['T'].values * units.degC
    RH = df['U'].values * units.percent
    
    q = mpcalc.mixing_ratio_from_relative_humidity(relative_humidity=RH,temperature=T,pressure=p)
    theta = mpcalc.potential_temperature(p, T)
    theta_v = mpcalc.virtual_potential_temperature(p,T,q)
    return q, theta, theta_v
    
```

```python
#---Plot
fig = plt.figure(figsize=(20,8))
left = -0.2
bottom = 1.1
alt_bnds = [0,3]

fnum = ['A','B','C','D']
labs_list = {'A':'a','B':'b','C':'c','D':'d'}

for idx,f in enumerate(fnum):
    df = read_radiosonde('Radiosonde-'+f+'.txt')
    q,theta,theta_v = check_calc(df)
    
    #---Plots
    ax = fig.add_subplot(1,4,idx+1)
    #ax.set_title('Water Vapor Mixing Ratio',loc='left')
    ax.set_xlim([0,10])
    ax.set_ylim(alt_bnds)
    ax.set_xlabel('Mixing ratio [g/kg]',c='dodgerblue')
    ax.grid()
    if idx == 0:
        ax.set_ylabel('Altitude [km]')

    #---Labels
    ax.text(left, bottom, labs_list[f],
             horizontalalignment='left',
             verticalalignment='bottom',
             weight='bold',
             fontsize='12',
             transform=ax.transAxes)

    ax.text(left+0.1, bottom, 'Radiosonde-'+str(f),
             horizontalalignment='left',
             verticalalignment='bottom',
             fontsize='12',
             transform=ax.transAxes)
    
    #---Mixing ratio
    ln1 = ax.plot(q*1000,df['Altitude']/1000, label='Mixing Ratio [g/kg]',
                 c='dodgerblue')
    
    #---Wind speed
    ax2 = ax.twiny()
    ax2.xaxis.set_ticks_position('bottom')
    rspine = ax2.spines['bottom']
    rspine.set_position(('axes', -0.08))
    ax2.set_xlabel('Wind speed [m/s]',c='fuchsia')
    ax2.xaxis.set_label_position('bottom')
    ax2.set_xlim([0,30])
    ax2.set_ylim(alt_bnds)
    
    ln2 = ax2.plot(df['Wsp'],df['Altitude']/1000, label='Wind Speed [m/s]',
                 c='fuchsia')
    
    #---Potential Temperature & Virtual Potential Temperature
    ax3 = ax.twiny()
    ax3.set_xlim([285,310])
    ax3.set_ylim(alt_bnds)
    ax3.set_xlabel('Potential Temperature [K]')
    
    ln3 = ax3.plot(theta,df['Altitude']/1000, label='Potential Temperature [K]',
                  c='red')
    ln4 = ax3.plot(theta_v,df['Altitude']/1000, label='Virtual Potential Temperature [K]',
                  c='green')

    #---Layer lines
    if idx == 0:
        ax.axhline(0.80,c='black',linestyle='--',linewidth=1)
        ax.text(1.04, 0.075, 'Surface Layer',
                ha='left',
                va='bottom',
                fontsize='12',
                transform=ax.transAxes)
        ax.axhline(2.45,c='black',linestyle='--',linewidth=1)
        ax.text(1.04, 0.5, 'Mixed Layer',
                ha='left',
                va='bottom',
                fontsize='12',
                transform=ax.transAxes)
        ax.text(1.04, 0.9, 'Free Troposphere',
             ha='left',
             va='bottom',
             fontsize='12',
             transform=ax.transAxes)
    elif idx == 1:
        ax.axhline(0.605,c='black',linestyle='--',linewidth=1)
        ax.text(1.04, 0.085, 'Stable Boundary Layer',
                ha='left',
                va='bottom',
                fontsize='12',
                transform=ax.transAxes)
        ax.text(1.04, 0.5, 'Free Troposphere',
                ha='left',
                va='bottom',
                fontsize='12',
                transform=ax.transAxes)
    elif idx == 2:
        ax.axhline(0.61,c='black',linestyle='--',linewidth=1)
    elif idx == 3:
        ax.axhline(0.365,c='black',linestyle='--',linewidth=1)
        ax.axhline(1.20,c='black',linestyle='--',linewidth=1)
        ax.text(1.04, 0.05, 'Surface Layer',
             ha='left',
             va='bottom',
             fontsize='12',
             transform=ax.transAxes)
        
    #---Legend
    if idx == 3:
        lns = ln1+ln2+ln3+ln4
        labs = [l.get_label() for l in lns]
        ax.legend(lns, labs, bbox_to_anchor=(1.1, 0.5), frameon=True)

#---Figure spacing & Title
fig.subplots_adjust(wspace=1.0)
fig.suptitle('4b) ', x=0.05, y=0.95,
             ha='left',
             va='bottom',
             fontsize='24');
```
